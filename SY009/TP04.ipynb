```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def add_labels(x, y, labels, ax=None):

    if ax is None:
        ax = plt.gca()
    for x, y, label in zip(x, y, labels):
        ax.annotate(label, [x, y], xytext=(10, -5), textcoords="offset points",)
    return ax
```

# Q1

```python
notes =pd.read_csv("TP04_Analyse_en_composantes_principales-enonce/data/notes.txt",sep="\s+")
cls = PCA(n_components=4)
pcs = cls.fit_transform(notes)
cls.components_
```

```python
plt.rcParams["figure.figsize"] = (5,3)
plt.bar(["Axe 1", "Axe 2", "Axe 3", "Axe 4"], cls.explained_variance_ratio_)
plt.show()
```

# Q2

```python
plt.scatter(pcs[:, 0], pcs[:, 1])
add_labels(pcs[:, 0], pcs[:, 1], notes.index)
plt.show()
```
